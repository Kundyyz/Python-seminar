#Задача №67. Решение в группах
1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
* high - длинный(от 42)
* middle - средний(от 35 до 42)
* low - маленький(до 35)

```python
# penguins[penguins['bill_length_mm'] > 42]
penguins.loc[penguins['bill_length_mm'] > 42, 'new_col'] = 'high'
```

```python
penguins
```

```python
penguins.loc[(penguins['bill_length_mm'] <= 42) & (penguins['bill_length_mm'] >= 35), 'new_col'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'new_col'] = 'low'
```
